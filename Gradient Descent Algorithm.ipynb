```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd

from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error
```

```python
x = np.arange(0, 20.1, 0.1)
np.random.seed(0)
y = 1*x**5 + 3*x**4 - 100*x**3 + 8*x**2 - (300*x) - 1e5 + np.random.randn(len(x))*1e5
```

```python
p = 4
n = len(x)
X = np.ones([n, p])

# creating polynomial
for i in range(n):
  for j in range(p):
    X[i][j] = x[i]**(j+1)

regression = LinearRegression().fit(X, y)

theta_0 = regression.intercept_*np.ones([n, 1])
theta = np.reshape(regression.coef_, (p, 1))
y_predicted = np.add(np.matmul(X, theta), theta_0)

rms_error = np.sqrt(mean_squared_error(y, y_predicted))       # RMS error calculation
print("RMS Error = ",rms_error)

fig = plt.figure()
ax1 = fig.add_subplot(111)
ax1.scatter(x, y, label='Y_Actual')
ax1.scatter(x, y_predicted, label='Y_Predicted')
plt.legend()
plt.show()

print("Theta Vector")
print(regression.intercept_, regression.coef_)
```

```python
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D

# Given data
X = np.array([1, 3, 6])
Y = np.array([6, 10, 16])

# Create a meshgrid of q0 and q1 values
q0_values = np.linspace(-100, 100, 100)
q1_values = np.linspace(-100, 100, 100)
q0_mesh, q1_mesh = np.meshgrid(q0_values, q1_values)

# Calculate RSS values for each combination of q0 and q1
rss_values = np.zeros_like(q0_mesh)
for i in range(len(q0_values)):
    for j in range(len(q1_values)):
        q0 = q0_values[i]
        q1 = q1_values[j]
        rss = np.sum((Y - (q0 + q1*X))**2)
        rss_values[j, i] = rss

# Create a colorful contour plot
plt.figure(figsize=(15, 5))
contour = plt.contourf(q0_mesh, q1_mesh, rss_values, levels=10, cmap='viridis')
plt.colorbar(contour, label='RSS')
plt.xlabel('Intercept (q0)')
plt.ylabel('Slope (q1)')
plt.title('Contour Plot of RSS in q0-q1 Space')
# plt.grid(True, linestyle='--', alpha=0.6)
plt.xticks(fontsize=10)
plt.yticks(fontsize=10)
plt.tight_layout()

plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
from matplotlib.animation import FuncAnimation

X = np.array([1, 3, 6])
Y = np.array([6, 10, 16])

def RSS(q0, q1):
    return np.sum((Y - (q0 + q1*X)) ** 2)
def gradient_descent(q0_init, q1_init, learning_rate, num_iterations):
    q0 = q0_init
    q1 = q1_init
    m = len(X)
    history = []
    for i in range(num_iterations):
        history.append((q0, q1))
        y_pred = q0 + q1 * X
        error = y_pred - Y
        grad_q0 = (2 / m) * np.sum(error)
        grad_q1 = (2 / m) * np.sum(error * X)
        q0 -= learning_rate * grad_q0
        q1 -= learning_rate * grad_q1
    return history

q0_init = 0
q1_init = 0
learning_rate = 0.01
num_iterations = 100

history = gradient_descent(q0_init, q1_init, learning_rate, num_iterations)
q0_grid = np.linspace(-100, 100, 100)
q1_grid = np.linspace(-100, 100, 100)

Q0_grid, Q1_grid = np.meshgrid(q0_grid, q1_grid)
Z_grid = np.zeros_like(Q0_grid)

for i in range(len(q0_grid)):
    for j in range(len(q1_grid)):
        Z_grid[i, j] = RSS(Q0_grid[i, j], Q1_grid[i, j])

fig, (ax1, ax2) = plt.subplots(1, 2)

# Contour plot
ax1.contour(Q0_grid, Q1_grid, Z_grid)
point = ax1.scatter([], [], color='red')
arrow = ax1.arrow([], [], [], [], color='red', head_width=0.5)

# Scatter plot with linear regression fit
ax2.scatter(X, Y)
line = ax2.plot([], [], color='red')[0]

def update(i):
    # Update contour plot
    point.set_offsets([history[i]])
    if i > 0:
        dx = history[i][0] - history[i-1][0]
        dy = history[i][1] - history[i-1][1]
        arrow.set_xy([[history[i-1][0], history[i-1][1]], [dx, dy]])
    
    # Update scatter plot with linear regression fit
    x_line = np.linspace(plt.xlim()[0], plt.xlim()[1], 100)
    y_line = history[i][0] + history[i][1] * x_line
    line.set_data(x_line, y_line)
    
    # Update title with iteration number and RSS value
    plt.suptitle(f'Iteration {i+1}, RSS = {RSS(*history[i]):.2f}')

anim = FuncAnimation(fig, update, frames=num_iterations, interval=200)
plt.show()
```
